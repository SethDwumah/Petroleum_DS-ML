# Fractional Flow Stimulation


### A clear depiction of how the fractional flow curve varies as we change the properties

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# sw = Water saturation
# so = Oil Saturation
# krw = permeability of water 
# kro = permeability of oil
sw = np.linspace(0.01,1,num=50)
```

```python
so = 1-sw
```

```python
krw = sw**3
```

```python
kro = (1 - sw)**3
```

```python
plt.style.use('default')
plt.plot(sw,krw)
plt.plot(sw,kro)
plt.xlabel('Water Saturation, Sw')
plt.ylabel('Relative Permeability, Krel')
plt.legend(['Krw','Kro'])
plt.grid()
```

```python
# Now Fraction flow part for a horizontal flow 

for i in range(3):
    ou = int(input("\nEnter oil viscosity: "))
    uw = int(input("\nEnter water viscosity: "))
    fw = 1/(1+(uw+kro)/(ou*krw))
    
    plt.plot(sw,fw)
    plt.xlabel('Water saturation, Sw')
    plt.ylabel('Fractional Flow of water, fw(bbl/day)')
    plt.legend(['1 CP water', '100 CP (polymer) water','1000 CP polymer Flood'])
    plt.grid()
```
