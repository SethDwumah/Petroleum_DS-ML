#Chapter 1: Solving ODEs

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.integrate import odeint

%matplotlib inline
```

#Problem 1


$ \frac{dy}{dx} = -ky $

```python
#1. Model creation

def model(y,x):
    k=0.5
    dydx=-k*y
    return dydx
# 2. x
x = np.linspace(0,10,20)

#3. Initial value
y0 = 5

# solution
y = odeint(model, y0,x)
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
```

#Problem 2


$ 5 \frac{dy(t)}{dt} = -y(t) + u(t) $

```python
def model(y,t):
    if t<=10:
        u=0
    else:
        u=10
    dydt = (-y +u)/5.0
    return dydt

t= np.linspace(0,100,100)
y0 = 10

y = odeint(model,y0,t)

plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('y')
plt.grid()
```

#Problem 3 Two ODE together


$ \frac{dx(t)}{dt} = -3exp(-t) $

$ \frac{dy(t)}{dt} = 3-y(t) $

```python
def modell(z,t):
    x = z[0]
    y = z[1]
    dxdt = -3.0*np.exp(-t)
    dydt = 3.0 -y
    
    return [dxdt,dydt]


z0 = [0,0]

t = np.linspace(0,10)

z = odeint(modell, z0,t)
```

```python

plt.plot(t,z[:,0], label= 'Solution for x(t)')
plt.plot(t,z[:,1], label='Solution for y(t)')

plt.grid()
plt.legend()
```
