## Columns - select input and output columns

#### Use Correlation Analysis as a Quick-Hack to do the same.

Too High and Too low Correlation magnitudes need to be avoided

```python
output_col = 'BORE_OIL_VOL'
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), annot=True, cmap='ocean')
```

```python
data.corr()[output_col].sort_values(ascending=False)[1:].plot(kind='bar',figsize=(12,4), grid=True)
```

```python
corr_df = pd.DataFrame(data.corr()[output_col].sort_values(ascending=False)[1:])
corr_df.columns = ['Corr']
```

```python
corr_df
```

## Input Columns : Selecting features with good (but not extreme) correlations with output variables

```python
input_cols = corr_df[(abs(corr_df['Corr']) > 0.2) & (corr_df['Corr']<0.90)].index
```

```python
input_cols
```

## Name Input matrix as X (Feature matrix) and Output matrix as y (Target )

```python
X = data[input_cols]
Y = data[output_col]
```

```python
X.shape,Y.shape
```

## Train-Test (or Validation) Split. 

In Time series Projects, we cannot do a shuffle/permutation based split. We need to maintain the sequence

Here, I am keeping last 291 days as Testing or Validation set.

```python

train_length = 3000 # days

x_train, y_train = X.iloc[:train_length,:] , Y[:train_length]
x_test , y_test = X.iloc[train_length:,:], Y[train_length:]
```

```python
x_train.shape , x_test.shape
```

```python
y_train.shape, y_test.shape
```

## Feature Scaling. Important for better Training (in select models).

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

## Option 1: Multiple Linear Regression Model

```python
    from sklearn.linear_model import LinearRegression
    
    lr = LinearRegression()
```

```python
lr.fit(x_train_scaled, y_train)
```

```python

y_pred = lr.predict(x_test_scaled)
```

```python
lr.score(x_test_scaled,y_pred)
```

```python
yp =lr.predict(x_train_scaled)
```
