```python
import seaborn as sns
import pandas as pd
import mstplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv("Shale_Gas_Wells_DataSet.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
plt.style.use('default')
fig = plt.figure(figsize=(6,5))
sns.displot(data['Well Spacing'], color='blue', kde=False)
plt.xlabel('Well Spacing')
plt.ylabel('Frequency')
plt.title('Distribution plot for Well Spacing')
plt.show()
```

```python
from scipy import stats


plt.style.use('default')
sns.jointplot(data['Stage Spacing'], data['EUR'], color = 'brown', kind='scatter')
```

```python
plt.style.use('default')
sns.jointplot(data['Proppant Loading'], data['EUR'], color='blue',kind='kde')
plt.show()
```

```python
plt.style.use('default')
sns.lmplot(x='Lateral Length',y='EUR', data=data)
```

```python
sns.boxplot(data['Porosity'], color='green')
```

```python
sns.kdeplot(data['Water Saturation'],data['EUR'],data=data,cmap='Reds',shade=True)
```
