```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Get the data

df_log = pd.read_csv('Sonic_Log_MIP_4H_DataSet.csv')
```

```python
fig,ax = plt.subplots(nrows=1, ncols=4, figsize=(16,12))
fig.suptitle('MIP_4H Sonic Log', fontsize=30)

Name = 'MIP_4H'

# Track #1
user1 = df_log[df_log['Well_Name']==Name]['GR_EDTC']
user2 = df_log[df_log['Well_Name']==Name]['DEPT']
ax1 = ax[0].twiny()
ax1.invert_yaxis()
ax1.plot(user1,user2, color='black', ls='-')
ax1.set_xlabel('GR_EDTC',color='black')
ax1.tick_params(axis='x',color='black')
ax1.spines['top'].set_position(('outward',1))

# Track #2
user3 = df_log[df_log['Well_Name']==Name]['NPHI']
user4 = df_log[df_log['Well_Name']==Name]['DEPT']
ax2 = ax[1].twiny()
ax2.invert_yaxis()
ax2.plot(user3,user4, color='red', ls='-')
ax2.set_xlabel('NPHI',color='red')
ax2.tick_params(axis='x',color='red')
ax2.spines['top'].set_position(('outward',1))

# Track #3
user5 = df_log[df_log['Well_Name']==Name]['RHOZ']
user6 = df_log[df_log['Well_Name']==Name]['DEPT']
ax3 = ax[2].twiny()
ax3.invert_yaxis()
ax3.plot(user5,user6, color='brown', ls='-')
ax3.set_xlabel('RHOZ',color='brown')
ax3.tick_params(axis='x',color='brown')
ax3.spines['top'].set_position(('outward',1))

# Track #4
ax4 = ax[3].twiny()
ax4.plot(df_log['PEFZ'],df_log['DEPT'], color='green', ls='-')
ax4.invert_yaxis()
ax4.set_xlabel('PEFZ',color='green')
ax4.tick_params(axis='x', color='green')
ax4.spines['top'].set_position(('outward',1))


plt.gca().invert_yaxis()
plt.interactive(True)
```
