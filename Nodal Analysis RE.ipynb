```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#  First lets get IPR dataset . {Source : Petrowiki}
 
q_ipr = [0,1999,3094,3902,4512,4963,5275,5458,5519]
pwf_ipr = [4000,3500,3000,2500,2000,1500,1000,500,14.7]

ipr = pd.DataFrame({'Pwf':pwf_ipr,
                   'Q':q_ipr})
```

```python
ipr
```

```python
# Lets visualize the data

x_ipr = q_ipr
y_ipr = pwf_ipr

plt.style.use('seaborn')
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,4))
plt.plot(x_ipr,y_ipr, label = "IPR")
plt.grid(color='white')
plt.xlabel('Qoil(bbl/d)', fontsize=15)
plt.ylabel('Pwf(psi)', fontsize=15)
plt.title('IPR Curve')

plt.xticks(size=10, color='black')
plt.yticks(size=10, color='black')
plt.show()
```

```python
# Flow TPR Data

# q = np.arange(1000,6500,500)

q_tpr = [1000,1500,2000,2500,3000,3500,4000,4500,5000,5500,6000]

p_190 = [1334,1400,1487,1592,1712,1843,1984,2132,2287,2446,2600]

p_2375 = [1298,1320,1351,1390,1435,1487,1545,1600,1677,1740,1824]

p_2875 = [1286,1294,1305,1319,1356,1336,1378,1403,1431,1461,1493]

tpr = pd.DataFrame({'q':q_tpr,'Pwf(1.90" tbg)':p_190,'Pwf(2.375" tbg)':p_2375,'Pwf(2.875" tbg)':p_2875})
tpr
```

```python
plt.figure(figsize=(8,4))

plt.style.use('seaborn')

plt.plot(ipr['Q'], ipr['Pwf'], label='IPR', linewidth=1.5)

plt.plot(tpr['q'], tpr['Pwf(1.90" tbg)'], label='TPR for 1.9" tubing', linewidth=1.5 )
plt.plot(tpr['q'], tpr['Pwf(2.375" tbg)'], label='TPR for 2.375" tubing', linewidth=1.5 )
plt.plot(tpr['q'], tpr['Pwf(2.875" tbg)'], label='TPR for 2.875" tubing', linewidth=1.5 )

plt.xlabel('FlowRate(Mscf/D)', fontsize=15)
plt.ylabel('Flowing BHP(Psia)', fontsize = 15)

plt.title('Nodal Analysis for Gas Well', fontsize=24)

plt.grid()

plt.legend(loc='best')

plt.show()
```
